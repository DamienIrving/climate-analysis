```python
# Import modules

import cdms2
import matplotlib.pyplot as plt
import os, sys
import numpy

cwd = os.getcwd()
repo_dir = '/'
for directory in cwd.split('/')[1:]:
    repo_dir = os.path.join(repo_dir, directory)
    if directory == 'phd':
        break

anal_dir = os.path.join(repo_dir, 'data_processing')
sys.path.append(anal_dir)
try:
    import calc_fourier_transform as cft
except ImportError:
    raise ImportError('Must run this script from anywhere within the phd git repo')
```

```python
# Read in data

fin = cdms2.open('/home/dbirving/Downloads/Data/va_Merra_250hPa_30day-runmean-2002_r360x181.nc')
data = fin('va', latitude=(-70,-40))
print data.shape
```

```python
# Calculate the amplitude spectrum

signal = data
indep_var = data.getLongitude()[:]

sig_fft, sample_freq = cft.fourier_transform(signal, indep_var)
print sig_fft.shape
print sample_freq.shape

amp = cft.spectrum(sig_fft, output='amplitude')
print amp.shape
```

```python
%matplotlib inline
plt.plot(sample_freq[10, 3, :])
plt.show()
```

```python
# Plot the power spectrum

def plot_spectrum(freqs, vals, tag=None, window=20):
    """Plot power spectrum.
    
    Input arguments:
        Window  -> There is an inset winow for the first 0:window frequencies
    
    """
    
    plt.figure()  
    
    # Outer plot
    plt.plot(freqs[1:window+1], vals[0:window])  # Because I think a freq of 0 makes no sense
    plt.xlabel('Frequency [cycles / domain]')
    plt.ylabel('amplitude')
    
    #outfile = 'power_spectrum_'+tag+'.png' if tag else 'power_spectrum.png'
    #plt.savefig(outfile)
    #plt.clf()
    plt.show()

print 'Singe day single latitude'
plot_spectrum(sample_freq[0, 0, :], amp[0, 0, :])

print 'Single day, latitudinal mean'
amp_lat_mean = numpy.mean(amp, axis=1)
print amp_mean.shape
plot_spectrum(sample_freq[0, 0, :], amp_lat_mean[0, :])


print 'Annual mean, latitudinal mean'
amp_annual_mean = numpy.mean(amp_lat_mean, axis=0)
print amp_annual_mean.shape
plot_spectrum(sample_freq[0, 0, :], amp_annual_mean)

```
