# Rossby wave source

### Background

The basic barotropic model to study Rossby wave propagation makes use of the vorticity equation at a single level in the upper troposphere. Following [Sardeshmukh (1988)]() and [James (1994)]() and neglecting friction, this can be written as

$$\frac{\partial \zeta}{\partial t} + \mathbf{v} \cdot \bigtriangledown \zeta = -\zeta D $$

where $\zeta (=f + \xi)$ is the absolute vorticity, $\xi (=\partial v / \partial x - \partial u / \partial y)$ is the vertical component of the relative vorticity, $f = 2\Omega \sin \phi$ is the Coriolis force, $t$ is time, $\mathbf{v} = (u,v)$ is the horizontal velocity, and $D (=\mathbf{v} \cdot \bigtriangledown )$ is the divergence. From continuity, $D = -\partial \omega / \partial p$.

The velocity field $\mathbf{v}$ may be decomposed into a purely rotational part $\mathbf{v}_\psi$ and a purely divergent part $\mathbf{v}_\chi$ (which is a general result known as Helmholtz's theorem). The velocity field can then be described in terms of two scalar fields, the streamfunction $\psi$ and the velocity potential $\chi$, since

$$\mathbf{v}_\psi = \mathbf{k} \times \bigtriangledown \psi,\;\;\;\;\; \mathbf{v}_\chi = \bigtriangledown \chi$$

From these definitions, it follows that the relative vorticity $\xi = \bigtriangledown^{2} \psi$ and the divergence $D = \bigtriangledown^{2} \chi$. Now substitute for $\mathbf{v}$ and re-write in the form 
  
$$\frac{\partial \zeta}{\partial t} + \mathbf{v}_\psi \cdot \bigtriangledown \zeta = -\zeta D - \mathbf{v}_\chi \cdot \bigtriangledown \zeta$$

This equation represents the correct partitioning between Rossby wave propagation terms (left), which involves only the rotational component, and forcing terms (right), with the divergent component. Analyzing this equation, the conclusion is that Rossby wave forcing is greater in areas where divergence, divergent wind, absolute vorticity and gradient of absolute vorticity are large. This can happen in regions poleward of diabatic tropical heating, where divergence at high levels, associated with deep convection, is greater and where there are large gradients of vorticity associated with jet streams. The western Pacific (to the east of Australia) presents such characteristics and is a favourable region for Rossby waves trains formation. Another favourable region with the dynamical conditions for Rossby wave formation is located at East Asia.

Forcing terms can be grouped in only one term called Rossby wave source ($S$): 

$$S = -\zeta D - \mathbf{v}_\chi \cdot \bigtriangledown \zeta = -\bigtriangledown \dot (\mathbf{v}_\chi \zeta)$$

This equation shows how Rossby waves can be excited by diabatic tropical heating (represented by divergence at high levels), although absolute vorticity is often small near the heating. Divergent flow is greater at the edge of the heating region, and not in the region where divergence is large, and the gradient of absolute vorticity is larger towards subtropics, implying that higher values of $S$ should be found in the subtropics [LachlanCope2006]().

It is this expression that is used to calculate the Rossby wave source on the [windspharm example page](http://ajdawson.github.io/windspharm/examples/rws_cdms.html). 

`S = -eta * div - uchi * etax + vchi * etay`

### Code validation

First, I define a function to calculate the Rossby wave source, based on the [windspharm example](http://ajdawson.github.io/windspharm/examples/rws_cdms.html):

```python
# Note that I'm using the IPython Notebook that comes with UV-CDAT 1.5.1 

%matplotlib inline

import numpy as np
import matplotlib as mpl
mpl.rcParams['mathtext.default'] = 'regular'
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import cdms2, MV2

from windspharm.cdms import VectorWind


def calc_wind_quantity(uwnd, vwnd, quantity):
    """Calculate some wind derived quanity from the zonal (uwnd) and meridional (vwnd) wind.
    
    quantity options:
        rws      ->  Rossby wave source
        rws1     ->  Vortex stretching term of the RWS
        rws2     ->  Advection of absolute vorticity the by divergent flow term of the RWS
    
    """
    
    assert quantity in ['rws', 'rws1', 'rws2']

    # Create a VectorWind instance to handle the computations.
    w = VectorWind(uwnd, vwnd)

    # Compute components of rossby wave source: absolute vorticity, divergence,
    # irrotational (divergent) wind components, gradients of absolute vorticity.
    eta = w.absolutevorticity()
    div = w.divergence()
    uchi, vchi = w.irrotationalcomponent()
    etax, etay = w.gradient(eta)

    # Combine the components to form the Rossby wave source term.
    if quantity == 'rws1':
        return -eta * div
    elif quantity == 'rws2':
        return -(uchi * etax + vchi * etay)
    elif quantity == 'rws' :
        return -eta * div - (uchi * etax + vchi * etay) ### THESE BRACKETS WERE FORGOTTEN 
                                                        ### ON THE WEBSITE EXAMPLE!!! 
    

def create_plot(data, title, units, scale_factor=1.0):
    m = Basemap(projection='cyl', resolution='c', llcrnrlon=0, llcrnrlat=-90,
                urcrnrlon=360.01, urcrnrlat=90)
    lons, lats = data.getLongitude()[:], data.getLatitude()[:]
    x, y = m(*np.meshgrid(lons, lats))
    clevs = [-30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30]
    m.contourf(x, y, data.asma()*scale_factor, clevs, cmap=plt.cm.RdBu_r,
               extend='both')
    m.drawcoastlines()
    m.drawparallels((-90, -60, -30, 0, 30, 60, 90), labels=[1,0,0,0])
    m.drawmeridians((0, 60, 120, 180, 240, 300, 360), labels=[0,0,0,1])
    plt.colorbar(orientation='horizontal')
    plt.title('%s (%s)' %(title, units), fontsize=14)
    plt.show()
```

I then re-create the example plot to make sure the code was copied correctly...

```python
# Read data
ncu = cdms2.open('uwnd_mean.nc', 'r')
uwnd = ncu('uwnd')
ncu.close()

ncv = cdms2.open('vwnd_mean.nc', 'r')
vwnd = ncv('vwnd')
ncv.close()

# Calculate Rossby wave source
rws = calc_wind_quantity(uwnd, vwnd, 'rws')
rws_dec = rws(time=slice(11,12), longitude=(0,360), squeeze=True)  #Pick Dec, add cyclic point for plotting

# Plot
title = 'Rossby wave source'
units = '$10^{-11}$s$^{-1}$'
create_plot(rws_dec, title, units, scale_factor=1e11)
```

### RWS components

In analysing the Rossby wave source [Shimizu2011]() divide $S$ into two terms (i.e. $S = S_1 + S_2$) and seasonal means were calculated from the daily values of each term. In particular, $S_1 = -\zeta D$ is the vortex stretching term, or the generation of vorticity by divergence, and $S_2 = -\mathbf{v}_\chi \cdot \bigtriangledown \zeta$ is the advection of absolute vorticity by divergent flow. These terms had been discussed previously by Lu and Kim (2004) in a different approach, focused on the Northern Hemisphere.

Unlike the example above, they do not calculate the Rossby wave source from the long term mean flow. In producing a climatology, they argue that you should calculate the Rossby wave source at every single daily time step (i.e. from the relevant values of absolute and relative vorticity, horizontal divergence and divergent wind), and then calculate the seasonal mean from this. This method of calculation has proved to be more appropriate than obtaining the seasonal means of Rossby wave source from the seasonal means of each component, as shown by [Simmonds2009](). According to these authors, to calculate such terms directly from the time-mean flow results in significant biases because these terms have strong temporal covariances.

We are interested here in reproducing the results of the climatology of [Shimizu2011](), to further validate the code.



```python
import cdutil

def plot_season(data, season, title, units, scale_factor):
    """Plot seasonal average from daily input data"""
    
    assert season in ['DJF', 'MAM', 'JJA', 'SON']
    cdutil.setTimeBoundsDaily(data, frequency=1)
    data_climatology = eval('cdutil.'+season+'.climatology(data)')
    data_climatology = data_climatology(longitude=(0,360), squeeze=1)

    create_plot(data_climatology, title+', '+season, units, scale_factor=scale_factor)
```

```python
# Read data
ncu = cdms2.open('ua_Merra_250hPa_daily-1990-1999_r144x73.nc', 'r')
uwnd = ncu('ua', time=('1990-01-01', '1991-12-31'))
ncu.close()

ncv = cdms2.open('va_Merra_250hPa_daily-1990-1999_r144x73.nc', 'r')
vwnd = ncv('va', time=('1990-01-01', '1991-12-31'))
ncv.close()
```

The results for the stretching term look very similar to Figure 5 of Shimizu2011...

```python
# Plot stretching term
rws1 = calc_wind_quantity(uwnd, vwnd, 'rws1')
units = '$10^{-11}$s$^{-1}$'
for seas in ['DJF', 'MAM', 'JJA', 'SON']:
    plot_season(rws1, seas, 'Stretching term', units, 1e11)
```

The results for the advection term look very similar to Figure 5 of Shimizu2011...

```python
# Plot advection term
rws2 = calc_wind_quantity(uwnd, vwnd, 'rws2')
units = '$10^{-11}$s$^{-1}$'
for seas in ['DJF', 'MAM', 'JJA', 'SON']:
    plot_season(rws2, seas, 'Advection term', units, 1e11)
```
