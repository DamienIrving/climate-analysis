```python
import pandas
```

```python
SAM_index_pandas = pandas.read_csv('/home/dbirving/Downloads/Data/newsam.1957.2014.txt', 
                                   skiprows=2, 
                                   delim_whitespace=True, 
                                   header=0, 
                                   index_col=0,
                                   names=range(1,13))
```

```python
SAM_index_pandas
```

```python
import cdtime

output_data = []
output_times = []
for year, data in SAM_index_pandas.iterrows():
    for month in data.keys():
        ct = cdtime.comptime(year, month, 15)
        rt = ct.torel("days since 1979")
        output_times.append(rt)
        output_data.append(data[month])
```
