## Index comparison

```python
quad_index_comparison = df[['ZW3_index', 'nenv_amp_mean', 'env_amp_mean', 'wave3_amp']]
quad_index_comparison.corr()
```

Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

```python
df['env_amp_mean_norm'] = normalise_series(df['env_amp_mean'])
df['nenv_amp_mean_norm'] = normalise_series(df['nenv_amp_mean'])
df['ZW3_index_norm'] = normalise_series(df['ZW3_index'])
df['wave3_amp_norm'] = normalise_series(df['wave3_amp'])
```

#### Find extrema

```python
from scipy import signal
```

Here's the local maxima on the curve... (I might need to play around with order, which determines how many surrounding grid points to consider)

```python
maxima_indexes = signal.argrelextrema(numpy.array(df['env_amp_mean']), numpy.greater, order=90)
df['env_amp_mean'][list(maxima_indexes)[0]]
```

Here's the location of the places where the env amplitude is much greater than the ZW3 index...

```python
df['diff_env_zw3'] = df['env_amp_mean_norm'] - df['ZW3_index_norm']
diff_env_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env_zw3']), numpy.greater, order=90)
df['diff_env_zw3'][list(diff_env_zw3_indexes)[0]]
```
