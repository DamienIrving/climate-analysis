It is interesting that the wave 3 component of the zg zonal anomaly correlates much more strongly with the va envelope (0.75) than the zg zonal anomaly envelope (0.45). Goes to show how dominant wave 1 is in the zg zonal anomaly data.

```python
from scipy import stats

stats.pearsonr(df_va['env_amp_mean'], df_zg['env_amp_mean'])[0]
```

Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

### Plotting with Pandas

#### Local maxima

(I might need to play around with order, which determines how many surrounding grid points to consider)

```python
from scipy import signal
```

```python
maxima_indexes_zg = signal.argrelextrema(numpy.array(df['env_amp_mean_zg']), numpy.greater, order=90)
print "Top ten local maxima (zg env mean)"
print df['env_amp_mean_zg'][list(maxima_indexes_zg)[0]].order(ascending=False)[0:10]
```

Some of the plots in this top 10 are really strong wave 1 patterns (e.g. 2004-08-13 and 1997-07-23), which the zg envelope appears to respond to. It turns out that the amplitude spectum for the zg zonal anomaly is way different than for va. Wave 1 is most dominant at all timescales for zg zonal anomaly, while for va wave 3 is most dominant.

To plot the envelope for these dates use:

```python
maxima_indexes_va_mean = signal.argrelextrema(numpy.array(df['env_amp_mean_va']), numpy.greater, order=90)
print "Top ten local maxima (va env mean)"
print df['env_amp_mean_va'][list(maxima_indexes_va_mean)[0]].order(ascending=False)[0:10]
```

This top 10 doesn't have the problem of strong wave 1 patterns, but there are times when very large values in a particular region make the overall index value a little larger than it ought to be (e.g. 1991-06-03, 1997-06-01, 2002-05-19). This might make a case for using the normalised envelope...

```python
maxima_indexes_va_mean_norm = signal.argrelextrema(numpy.array(df['nenv_amp_mean_va']), numpy.greater, order=90)
print "Top ten local maxima (va nenv mean)"
print df['nenv_amp_mean_va'][list(maxima_indexes_va_mean_norm)[0]].order(ascending=False)[0:10]
```

But the normalisation does some weird things in some regions, so a safer bet might be to simply take the median of the raw envelope...

```python
maxima_indexes_va_median = signal.argrelextrema(numpy.array(df['env_amp_median']), numpy.greater, order=90)
print "Top ten local maxima (va env median)"
print df['env_amp_median'][list(maxima_indexes_va_median)[0]].order(ascending=False)[0:10]
```

The median makes a small difference, but it's not compelling. I'd be justified in using either env-va-mean or env-va-median, but probably not nenv or zg.

Also interesting to consider the top 10 ZW3 index points:

```python
maxima_indexes_zw3 = signal.argrelextrema(numpy.array(df['ZW3_index_va']), numpy.greater, order=90)
print "Top ten local maxima (ZW3 index)"
print df['ZW3_index_va'][list(maxima_indexes_zw3)[0]].order(ascending=False)[0:10]
```
