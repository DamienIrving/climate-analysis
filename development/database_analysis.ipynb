It is interesting that the wave 3 component of the zg zonal anomaly correlates much more strongly with the va envelope (0.75) than the zg zonal anomaly envelope (0.45). Goes to show how dominant wave 1 is in the zg zonal anomaly data.

```python
from scipy import stats

stats.pearsonr(df_va['env_amp_mean'], df_zg['env_amp_mean'])[0]
```

Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

Still not super complelling. I need a better way of showing that my metric is better. Part of the problem is that they are a mixing pattern with intensity, so strong single point can take over.

Clearly I can show that the pattern moves around enough to make those single points not great. The question is, are the limitations of my metric less of a problem than that issue? (i.e. because no metric is perfect)

One idea might be to take the top 100 of each and calculate the average of the other, to see which is responding better? Or create a scatter plot...

```python
# Initialise figure
plt.figure(figsize=(10, 10))
plt.axhline(y=0, linestyle='-', color='0.5')
plt.axvline(x=0, linestyle='-', color='0.5')

x = df['env_amp_median_norm_va'][::3]
y = df['ZW3_index_norm'][::3]
c = df_va['wave3_phase_middle'][::3]

# Line of best fit
p = numpy.polyfit(x, y, 1)
print p

# Plot
plt.scatter(x, y, c=c, cmap='jet')
plt.plot(x, p[0]*x+p[1], 'r')
plt.xlabel("My metric")
plt.ylabel("ZW3 index")

plt.show()
```

It can be seen that there is far more spread from the origin (0, 0) in the bottom right quadrant (my metric large, ZW3 small) than there is in the top left (my metric small, ZW3 large). This means it is more common for the ZW3 metric to miss a large value of mine than vice-versa.

We can also see that the colors (the phase of the wave 3 component) are not randomly distributed. All the low ZW3 index values occur when the phase isn't aligned with the location of three ZW3 points.  

```python
bad_ZW3 = df[(df['env_amp_median_norm_va'] > 1) & (df['ZW3_index_norm'] < -3)]
print bad_ZW3['ZW3_index_norm']
```
