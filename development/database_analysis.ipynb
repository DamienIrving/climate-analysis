## Index comparison

```python
index_comparison = df[['ZW3_index_va', 'nenv_amp_mean_va', 'env_amp_mean_va', 'wave3_amp_max_va', 'nenv_amp_mean_zg', 'env_amp_mean_zg', 'wave3_amp_max_zg']]
index_comparison.corr()
```

It is interesting that the wave 3 component of the zg zonal anomaly correlates much more strongly with the va envelope (0.75) than the zg zonal anomaly envelope (0.45). Goes to show how dominant wave 1 is in the zg zonal anomaly data.

```python
from scipy import stats

stats.pearsonr(df_va['env_amp_mean'], df_zg['env_amp_mean'])[0]
```

Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

### Plotting with Pandas

```python
df['env_amp_mean_norm_va'] = normalise_series(df['env_amp_mean_va'])
df['nenv_amp_mean_norm_va'] = normalise_series(df['nenv_amp_mean_va'])
df['ZW3_index_norm'] = normalise_series(df['ZW3_index_va'])
df['wave3_amp_max_norm_va'] = normalise_series(df['wave3_amp_max_va'])

df['env_amp_mean_norm_zg'] = normalise_series(df['env_amp_mean_zg'])
df['nenv_amp_mean_norm_zg'] = normalise_series(df['nenv_amp_mean_zg'])
df['wave3_amp_max_norm_zg'] = normalise_series(df['wave3_amp_max_zg'])
```

#### Local maxima

(I might need to play around with order, which determines how many surrounding grid points to consider)

```python
from scipy import signal
```

```python
maxima_indexes_zg = signal.argrelextrema(numpy.array(df['env_amp_mean_zg']), numpy.greater, order=90)
print "Top ten local maxima (zg env)"
print df['env_amp_mean_zg'][list(maxima_indexes)[0]].order(ascending=False)[0:10]
```

Some of the plots in this top 10 are really strong wave 1 patterns (e.g. 2004-08-13 and 1997-07-23), which the zg envelope appears to respond to. It turns out that the amplitude spectum for the zg zonal anomaly is way different than for va. Wave 1 is most dominant at all timescales for zg zonal anomaly, while for va wave 3 is most dominant.

To plot the envelope for these dates use:

```python
maxima_indexes_va = signal.argrelextrema(numpy.array(df['env_amp_mean_va']), numpy.greater, order=90)
print "Top ten local maxima (va env)"
print df['env_amp_mean_va'][list(maxima_indexes)[0]].order(ascending=False)[0:10]
```

This top 10 doesn't have the problem of strong wave 1 patterns, but there are times when very large values in a particular region make the overall index value much larger than it ought to be (e.g. 1995-07-17). This might make a case for using the normalised envelope...

```python
maxima_indexes_va_norm = signal.argrelextrema(numpy.array(df['nenv_amp_mean_va']), numpy.greater, order=90)
print "Top ten local maxima (va nenv)"
print df['nenv_amp_mean_va'][list(maxima_indexes)[0]].order(ascending=False)[0:10]
```

Although 1995-07-17 is in this list too, so the normalisation might not be having the desired effect...

#### Difference between indices

env versus ZW3

```python
# env vs. zw3
df['diff_env-va_zw3'] = df['env_amp_mean_norm_va'] - df['ZW3_index_norm']

start_index = list(df.index).index('1986-01-01')
end_index = list(df.index).index('1986-12-31')

df['diff_env-va_zw3'][start_index:end_index].plot()
```

```python
# Find the points of largest difference
diff_envva_gt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env-va_zw3']), numpy.greater, order=90)
print 'Top 10, env_va > zw3'
print df['diff_env-va_zw3'][list(diff_envva_gt_zw3_indexes)[0]].order(ascending=False)[0:10]

diff_envva_lt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env-va_zw3']), numpy.less, order=90)
print 'Top 10, env_va < zw3'
print df['diff_env-va_zw3'][list(diff_envva_lt_zw3_indexes)[0]].order(ascending=True)[0:10]
```

Interestingly most of the env > zw3 are in the winter months and env < zw3 in the summer months...

env versus wave 3

```python
# env vs. wave3

df['diff_env_wave3_va'] = df['env_amp_mean_norm_va'] - df['wave3_amp_max_norm_va']

start_index = list(df.index).index('1989-01-01')
end_index = list(df.index).index('1989-12-31')

df['diff_env_wave3_va'][start_index:end_index].plot()
```

```python
# Find the points of largest difference
diff_env_gt_wave3_indexes = signal.argrelextrema(numpy.array(df['diff_env_wave3_va']), numpy.greater, order=90)
print 'Top 10, env > wave3'
print df['diff_env_wave3_va'][list(diff_env_gt_wave3_indexes)[0]].order(ascending=False)[0:10]

diff_env_lt_wave3_indexes = signal.argrelextrema(numpy.array(df['diff_env_wave3_va']), numpy.less, order=90)
print 'Top 10, env < wave3'
print df['diff_env_wave3_va'][list(diff_env_lt_wave3_indexes)[0]].order(ascending=True)[0:10]
```
