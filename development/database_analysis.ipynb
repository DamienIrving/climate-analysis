## Index comparison

```python
quad_index_comparison = df[['ZW3_index', 'nenv_amp_mean', 'env_amp_mean', 'wave3_amp_max']]
quad_index_comparison.corr()
```

Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

```python
df['env_amp_mean_norm'] = normalise_series(df['env_amp_mean'])
df['nenv_amp_mean_norm'] = normalise_series(df['nenv_amp_mean'])
df['ZW3_index_norm'] = normalise_series(df['ZW3_index'])
df['wave3_amp_max_norm'] = normalise_series(df['wave3_amp_max'])
```

### Plotting with Pandas

```python
columns_for_plot = df[['env_amp_mean_norm', 'nenv_amp_mean_norm', 'ZW3_index_norm', 'wave3_amp_max_norm']]

start_index = list(df.index).index('1989-01-01')
end_index = list(df.index).index('1989-12-31')

columns_for_plot[start_index:end_index].plot()
```

Here's the local maxima on the curve... (I might need to play around with order, which determines how many surrounding grid points to consider)

```python
from scipy import signal
```

```python
maxima_indexes = signal.argrelextrema(numpy.array(df['env_amp_mean']), numpy.greater, order=90)
df['env_amp_mean'][list(maxima_indexes)[0]].order(ascending=False)[0:10]
```

#### Difference between indices

env versus ZW3

```python
# env vs. zw3
df['diff_env_zw3'] = df['env_amp_mean_norm'] - df['ZW3_index_norm']

start_index = list(df.index).index('1986-01-01')
end_index = list(df.index).index('1986-12-31')

df['diff_env_zw3'][start_index:end_index].plot()
```

```python
# Find the points of largest difference
diff_env_gt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env_zw3']), numpy.greater, order=90)
print 'env > zw3'
print df['diff_env_zw3'][list(diff_env_gt_zw3_indexes)[0]].order(ascending=False)[0:10]

diff_env_lt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env_zw3']), numpy.less, order=90)
print 'env < zw3'
print df['diff_env_zw3'][list(diff_env_lt_zw3_indexes)[0]].order(ascending=True)[0:10]
```

env versus wave 3

```python
# env vs. wave3

df['diff_env_wave3'] = df['env_amp_mean_norm'] - df['wave3_amp_max_norm']

start_index = list(df.index).index('1989-01-01')
end_index = list(df.index).index('1989-12-31')

df['diff_env_wave3'][start_index:end_index].plot()
```

```python
# Find the points of largest difference
diff_env_gt_wave3_indexes = signal.argrelextrema(numpy.array(df['diff_env_wave3']), numpy.greater, order=90)
print 'env > wave3'
print df['diff_env_wave3'][list(diff_env_gt_wave3_indexes)[0]].order(ascending=False)[0:10]

diff_env_lt_wave3_indexes = signal.argrelextrema(numpy.array(df['diff_env_wave3']), numpy.less, order=90)
print 'env < wave3'
print df['diff_env_wave3'][list(diff_env_lt_wave3_indexes)[0]].order(ascending=True)[0:10]
```
