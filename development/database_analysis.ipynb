Keep in mind that since we are really only interested in the large values of these indices, some of the discrepancy in correlation might arise in mid-to-low values, which we don't really care about so much.

For ease of comparsion, we should go ahead and normalise these indices so that they can be compared on the same plot...

```python
def normalise_series(series):
    """Normalise a Pandas data series by subtracting the
    mean and dividing by the stdev"""

    ave = series.mean()
    stdev = series.std()
    
    return (series - ave) / stdev
```

```python
df_va['env_amp_mean_norm'] = normalise_series(df_va['env_amp_mean'])
df_va['nenv_amp_mean_norm'] = normalise_series(df_va['nenv_amp_mean'])
df_va['ZW3_index_norm'] = normalise_series(df_va['ZW3_index'])
df_va['wave3_amp_max_norm'] = normalise_series(df_va['wave3_amp_max'])

df_zg['env_amp_mean_norm'] = normalise_series(df_zg['env_amp_mean'])
df_zg['nenv_amp_mean_norm'] = normalise_series(df_zg['nenv_amp_mean'])
df_zg['ZW3_index_norm'] = normalise_series(df_zg['ZW3_index'])
df_zg['wave3_amp_max_norm'] = normalise_series(df_zg['wave3_amp_max'])
```

#### Difference between indices

env versus ZW3

```python
# env vs. zw3
df['diff_env_zw3'] = df['env_amp_mean_norm'] - df['ZW3_index_norm']

start_index = list(df.index).index('1986-01-01')
end_index = list(df.index).index('1986-12-31')

df['diff_env_zw3'][start_index:end_index].plot()
```

```python
# Find the points of largest difference
diff_env_gt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env_zw3']), numpy.greater, order=90)
print 'Top 10, env > zw3'
print df['diff_env_zw3'][list(diff_env_gt_zw3_indexes)[0]].order(ascending=False)[0:10]

diff_env_lt_zw3_indexes = signal.argrelextrema(numpy.array(df['diff_env_zw3']), numpy.less, order=90)
print 'Top 10, env < zw3'
print df['diff_env_zw3'][list(diff_env_lt_zw3_indexes)[0]].order(ascending=True)[0:10]
```
