```python
from scipy import fftpack
from scipy import signal

import numpy
import math

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fourier_transform(signal, indep_var):
    """Calculate the Fourier Transform.
    
    Input arguments:
        indep_var  ->  Independent variable (i.e. 1 dimensional time axis or longitude axis)
    
    Output:
        sig_fft    ->  Coefficients obtained from the Fourier Transform
        freqs      ->  Wave frequency associated with each coefficient
    
    """
    
    spacing = indep_var[1] - indep_var[0]
    sig_fft = fftpack.fft(signal)
    sample_freq = fftpack.fftfreq(len(indep_var), d=spacing) * len(indep_var) * spacing  # i.e. in units of cycles per length of domain
    sample_freq = numpy.resize(sample_freq, sig_fft.shape)
    
    return sig_fft, sample_freq
```

```python
def spectrum(signal_fft, scaling='amplitude', variance=None):
    """Calculate the spectral density for a given Fourier Transform.
    
    The choices for the amplitude scaling for each frequency
    are as follows (see Wilks 2011, p440):
    - 'amplitude' : no scaling at all (C)
    - 'power' : sqaure the amplitude (C^2)
    - 'R2' : variance explained = [(n/2)*C^2] / (n-1)*variance^2, 
      where n and variance are the length and variance of the 
      orignal data series
      (R2 = the proportion of the variance explained by each harmonic)    

    The sample frequencies include both the positive
    and negative frequencies (which are identical for real functions
    and maybe other times as well), so when plotting the amplitude
    or R2 you can just double the positive value 
    
    """
    
    assert scaling in ['amplitude', 'power', 'R2']
    if scaling == 'R2':
        assert variance, \
        "To calculate variance explained must provide variance value" 
    
    n = len(signal_fft)
    amp = numpy.abs(signal_fft) / n
    if scaling == 'amplitude':
        result = amp
    elif scaling == 'power':
        result = (amp)**2
    elif scaling == 'R2':
        result = ((n / 2) * ((amp*2)**2)) / ((n - 1) * (variance))
    
    return result
```

```python
# Example signal

t = numpy.linspace(0, 1, 100)
y = 8*numpy.sin((2 * math.pi * t * 4) - numpy.deg2rad(45)) + numpy.sin(2 *math.pi * t *10) + numpy.sin(2 * math.pi * t * 13)

plt.figure()

# Plot signal
plt.plot(t, y, 'b-')
plt.xlabel("TIME (sec)")
plt.ylabel("ORIGINAL SIGNAL")
plt.show()
```

```python
# Fourier transform
sig_fft, sample_freq = fourier_transform(y, t)

plt.figure()
plt.plot(sample_freq, numpy.abs(sig_fft), 'b-')
plt.xlabel("FREQ")
plt.ylabel("AMPLITUDE")
plt.show()
```

```python
# let's take only the positive frequencies and normalize the amplitude
N = len(sig_fft)
yfft  = numpy.abs(sig_fft)/N
#freqs = freqs[:np.floor(N/2)]
#yfft  = yfft[:np.floor(N/2)]

plt.figure()
plt.plot(sample_freq[:numpy.floor(N/2)], yfft[:numpy.floor(N/2)], 'b-')
plt.xlabel("FREQ")
plt.ylabel("AMPLITUDE")
plt.show()
```

```python
# Calculate spectrum my way

variance = numpy.var(y)

amp_spectrum = spectrum(sig_fft, scaling='amplitude', variance=None)
power_spectrum = spectrum(sig_fft, scaling='power', variance=None)
r2_spectrum = spectrum(sig_fft, scaling='R2', variance=variance)
```

```python
print "Amplitude spectrum"
plt.figure()
plt.plot(sample_freq, amp_spectrum, 'b-')
plt.xlabel("FREQ")
plt.ylabel("AMPLITUDE")
plt.show()
```

```python
print "Power spectrum"
plt.figure()
plt.plot(sample_freq, power_spectrum, 'b-')
plt.xlabel("FREQ")
plt.ylabel("POWER")
plt.show()
```

```python
print "Variance explained spectrum"
plt.figure()
plt.plot(sample_freq, r2_spectrum, 'b-')
plt.xlabel("FREQ")
plt.ylabel("VARIANCE EXPLAINED")
plt.show()
```
