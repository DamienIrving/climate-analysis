```python
import math
import numpy
from scipy import fftpack
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
```

# The proper method

The most intuitive way to get the phase information out of a Fourier Transform (which gives you a magnitude and phase coefficient for each frequency) would be to get it from the transform itself. In testing whether this was a good way to go, I basically varied the resolution of the time axis, t.

```python
# Construct signal 
t = numpy.linspace(0, 1, 100000)
y = numpy.sin((2 * math.pi * t * 4) - numpy.deg2rad(45))  #+ numpy.sin(2 *math.pi * t *10) + numpy.sin(2 * math.pi * t * 13)
```

```python
plt.figure(figsize=(6,12))

# Plot signal
plt.subplot(411)
plt.plot(t, y, 'b-')
plt.xlabel("TIME (sec)")
plt.ylabel("ORIGINAL SIGNAL")

# Compute FFT 
F = fftpack.fft(y)
N = len(t)             # number of samples
dt = 0.00001             # inter-sample time difference
w = fftpack.fftfreq(N, dt)     # gives us a list of frequencies for the FFT
ipos = numpy.where(w > 0)
freqs = w[ipos]        # only look at positive frequencies
mags = numpy.abs(F[ipos])    # magnitude component
phase = numpy.imag(F[ipos])  # phase component

# Plot magnitude spectrum
plt.subplot(412)
plt.plot(freqs, mags, 'b-')
plt.xlim([0, 50])          # replot but zoom in on freqs 0-50 Hz
plt.ylabel("FFT MAGNITUDE")
plt.xlabel("FREQUENCY (Hz)")

# Plot phase spectrum
plt.subplot(413)
plt.plot(freqs, phase, 'b-')
plt.xlim([0, 50])          # replot but zoom in on freqs 0-50 Hz
plt.ylabel("FFT PHASE")
plt.xlabel("FREQUENCY (Hz)")

# Plot reconstructed signal
plt.subplot(414)
yr = fftpack.ifft(F)
plt.plot(t, y, 'b-')
plt.plot(t, yr, 'r-')
plt.legend(("original","reconstructed"))
plt.xlabel("TIME (sec)")
plt.ylabel("RECONSTRUCTED SIGNAL")

plt.show()
```

```python
print mags.min(), mags.max()
print mags[3]
print 'phase angle:', numpy.rad2deg(numpy.angle(F[ipos]))[3]
```

**Conclusion:** The phase angle can be quite imprecise for low resolution time axes (like my data). An alternative might be better. 

# Local maxima approach

An alternative would be to simply find the first local maxima.

```python
# Example for finding local maxima
x = numpy.array([2, 1, 2, 3, 2, 0, 1, 0])
lons = numpy.array([0, 45, 90, 135, 180, 225, 270, 315])
print signal.argrelextrema(x, numpy.greater)
print signal.argrelextrema(x, numpy.less)
```

I could then just do a scatterplot to look for clustering (wave 1 and 3 should cluster, others should be random). Keeping in mind that the x-axis wraps around.

An objective measure of the spread or phase shift (accounting for the wrap around) would be a little more difficult. I'd basically need to calculate the average distance between the location of the first local maxima at each timestep with every other timestep. The cross correlation (`scipy.signal.signaltools.correlate(A, B)`) might also be useful for looking at phase shift.

```python
big_x = numpy.resize(x, [2, 3, 8])
print big_x.shape
print big_x[0,2,:]
print big_x[1,2,:]
```

```python
print signal.argrelextrema(big_x, numpy.greater, axis=-1)
```

```python
def get_first_localmax(data):
    localmax_indexes = signal.argrelextrema(data, numpy.greater)
    
    return localmax_indexes[0][0]
    
```

```python
test_indexes = numpy.apply_along_axis(get_first_localmax, -1, big_x)
```

```python
print test_indexes.shape
print test_indexes
```

```python
# numpy.vectorize might also be an option here...

test_lons = map(lambda x: lons[x], test_indexes)
```
