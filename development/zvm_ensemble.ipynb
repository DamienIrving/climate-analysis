```python
import iris
import iris.plot as iplt
import matplotlib.pyplot as plt
import numpy
from iris.experimental.equalise_cubes import equalise_attributes
```

```python
%matplotlib inline
```

```python
infile1 = '/g/data/r87/dbi599/drstree/CMIP5/GCM/CCCMA/CanESM2/historical/yr/ocean/thetao-maps/ensmean-i1p1/thetao-maps-time-trend_Oyr_CanESM2_historical_ensmean-i1p1_1950-01-01_2000-12-31.nc'
infile2 = '/g/data/r87/dbi599/drstree/CMIP5/GCM/NCAR/CCSM4/historical/yr/ocean/thetao-maps/ensmean-i1p1/thetao-maps-time-trend_Oyr_CCSM4_historical_ensmean-i1p1_1950-01-01_2000-12-31.nc'
```

```python
var = 'zonal vertical mean globe argo sea water potential temperature'
```

```python
cube1 = iris.load_cube(infile1, var)
cube2 = iris.load_cube(infile2, var)
```

```python
print cube1
```

```python
print cube2
```

```python
iplt.plot(cube1, color='red', alpha=0.4, label='CMIP5 models')
iplt.plot(cube2, color='red', alpha=0.4)
plt.legend()
plt.show()
```

```python
print cube1.coord('latitude').points
```

```python
sample_points = [('latitude', numpy.arange(-90, 90, 1))]
new_cube1 = cube1.interpolate(sample_points, iris.analysis.Linear())
new_cube2 = cube2.interpolate(sample_points, iris.analysis.Linear())
```

```python
print new_cube1
```

```python
iplt.plot(cube1, color='red', alpha=0.4)
iplt.plot(new_cube1, color='blue', alpha=0.4)
iplt.plot(cube2, color='red')
iplt.plot(new_cube2, color='blue')
plt.show()
```

```python
new_cube1.add_aux_coord(iris.coords.DimCoord(1, 'realization'))
new_cube2.add_aux_coord(iris.coords.DimCoord(2, 'realization'))
```

```python
print new_cube1
```

```python
cubes = iris.cube.CubeList([new_cube1, new_cube2])
equalise_attributes(cubes)
merged_cube = cubes.merge_cube()
```

```python
print merged_cube
```

```python
ensemble_mean = merged_cube.collapsed('realization', iris.analysis.MEAN)
```

```python
print ensemble_mean
```

```python
iplt.plot(new_cube1, color='blue', alpha=0.4)
iplt.plot(new_cube2, color='blue')
iplt.plot(ensemble_mean, color='black')
plt.show()
```
